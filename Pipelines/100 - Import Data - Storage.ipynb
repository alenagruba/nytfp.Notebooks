```
# This notebook describes how to load data from Azure Storage
```

```
storage_account_name = "devmsbistor"
storage_account_access_key = "0whDp****"
container_name = "srcdata"
f_name = "kyphosis.csv"

file_location = "wasbs://{}@{}.blob.core.windows.net/{}".format(container_name, storage_account_name, f_name)

file_type = "csv"

spark.conf.set("fs.azure.account.key."+storage_account_name+".blob.core.windows.net", storage_account_access_key)
```

```
df = spark.read.format(file_type).option("inferSchema", "true").option("header","true").load(file_location)
```
