```
file_location = "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv"
file_type = "csv"
df = spark.read.format(file_type).option("inferSchema", "true").option("header", "true").load(file_location) 
df.count()
```

```
# Get df columns
df.columns
```

```
# Show one df column
display(df.select("price"))
```

If you want to query this df as a table, you can simply register it as a view or a table.

### I. Create a View using Python

```
# Create a View using Python
df.createOrReplaceTempView("V_Diamonds")
```

```
%sql
SELECT price FROM V_Diamonds
```

### II. Create a table using Python

```
# mode :
# - overwrite (same as drop & create) 
# - append (data will be added to the end of table)
df.write.format("parquet").mode("overwrite").saveAsTable("diamonds_python")
```

```
%sql
SELECT price
FROM diamonds_python
LIMIT 5
```

### III. Create a table using SQL

```
%sql
-- We can use SQL to create a table
DROP TABLE IF EXISTS diamonds;

CREATE TABLE diamonds
  USING csv
  OPTIONS (path "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header "true")
```

```
%sql
SELECT price
FROM diamonds
```
