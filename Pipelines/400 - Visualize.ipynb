```
df_pd = df.toPandas()
df_pd_eda = df_pd[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price']].copy()

gg = sns.pairplot(df_pd_eda, hue='color', palette='deep')
display(gg.fig)
```

```
sns.set(style="white")

g = sns.PairGrid(df_pd_eda, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3)

g.map_upper(sns.regplot)

display(g.fig)
```
